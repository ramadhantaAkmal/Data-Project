```python
import nltk
import re
import string
```

```python
sa = pd.read_csv('Gojek_CepetanGoSendInstant by Ariel NOAH x GoSend_comments.csv')
sa
```

```python
columns_keep = ["Comment"]
columns_useless = [x for x in sa if x not in columns_keep]
```

```python
sa_com= sa.drop(columns_useless,axis=1)
sa_com
```

```python
sa_com_new = sa_com.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))
sa_com_new
```

```python
def cleaningText(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) # menghilangkan mentions
    text = re.sub(r'#[A-Za-z0-9]+', '', text) # menghilangkan hashtag
    text = re.sub(r"http\S+", '', text) # menghilangkan link
    text = re.sub(r'[0-9]+', '', text) # menghilangkan numbers

    text = text.replace('\n', ' ') # menggantikan garis baru menjadi spasi
    text = text.translate(str.maketrans('', '', string.punctuation)) # menghilangkan semua tanda baca
    text = text.strip(' ') # menghilangkan whitespace
    return text
def casefoldingText(text): # merubah semua huruf menjadi huruf kecil
    text = text.lower() 
    return text
def tokenizingText(text): # mensplit text menjadi list token
    text = word_tokenize(text) 
    return text

def filteringText(text): # menghilangkan stopwords dari text
    listStopwords = set(stopwords.words('indonesian'))
    filtered = []
    for txt in text:
        if txt not in listStopwords:
            filtered.append(txt)
    text = filtered 
    return text

def stemmingText(text): # mengubah kalimat menjadi kalimat netral
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    text = [stemmer.stem(word) for word in text]
    return text

def toSentence(list_words): 
    sentence = ' '.join(word for word in list_words)
    return sentence
```

```python
sa_com_new['text_clean'] = sa_com_new['Comment'].apply(cleaningText)
sa_com_new['text_clean'] = sa_com_new['text_clean'].apply(casefoldingText)
sa_com_new.drop(['Comment'], axis = 1, inplace = True)

sa_com_new['text_preprocessed'] = sa_com_new['text_clean'].apply(tokenizingText)
sa_com_new['text_preprocessed'] = sa_com_new['text_preprocessed'].apply(filteringText)
sa_com_new['text_preprocessed'] = sa_com_new['text_preprocessed'].apply(stemmingText)

#menghapus kata yang duplikat
sa_com_new.drop_duplicates(subset = 'text_clean', inplace = True)
```

```python
df=sa_com_new.reset_index(drop=True)
df
```

```python
text=sa_com_new.iloc[:,1]
```

```python
a=[]
b=[]
for tes in text:
  for tos in tes:
    b.append(GoogleTranslator(source='auto', target='en').translate(tos))
  a.append(b)
  b=[]
```

```python
a
```

```python
trans = pd.DataFrame(np.array([a]).T)
trans.columns =['Translated']
```

```python
trans
```

```python
polarity=[]
pol=''
tot=0
for bas in trans['Translated']:
  for bes in bas:
    Text=TextBlob(bes)
    tot=tot+Text.sentiment.polarity
  if tot<0:
    pol='Negatif'
  elif tot>0:
    pol='Positif'
  else:
    pol='Netral'
  tot=0
  polarity.append(pol)
```

```python
df['Polarity']= pd.DataFrame(np.array([polarity]).T)
```
